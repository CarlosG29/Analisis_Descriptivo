```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Cargar los datos desde archivo CSV
df = pd.read_csv("train.csv")

# Mostrar las primeras 5 filas del dataset
print(df.head())

# Descripcion general del dataset
print(df.info())  # Informaci√≥n sobre las columnas y tipos de datos
```

```python

# Estadssticas descriptivas de las variables numericas
print(df.describe())

# Tabla de frecuencias para la columna Product Name
product_freq = df['Product Name'].value_counts()
print(product_freq.head())
```

```python
# Seleccionar las 5 ciudades con mayores ventas totales
top_10_cities = df.groupby('City')['Sales'].sum().nlargest(10)
print(top_10_cities)

# Seleccionar las 5 personas con mayores ventas totales
top_10_customers = df.groupby('Customer Name')['Sales'].sum().nlargest(5)
print(top_10_customers)
```

```python
import matplotlib.pyplot as plt
# grafico de barras para las 5 ciudades con mayores ventas totales
top_10_cities.plot(kind='bar', title='Top 10 Ciudades con Mayores Ventas')
plt.ylabel('Ventas Totales')
plt.xlabel('Ciudad')
plt.xticks(rotation=60)
plt.show()
```

```python
import matplotlib.pyplot as plt

# El histograma muestra la distribuci√≥n de las ventas (`Sales`) en todo el dataset, lo que nos ayuda a entender c√≥mo est√°n distribuidos los datos.

plt.hist(df['Sales'], bins=20, color='yellow', edgecolor='black', linewidth=1.2)

# Ajustar el rango del eje X para mostrar numeros mas peque√±os
plt.xlim(0, 4000)  
plt.ylim(0, 10000)

# Configurar el t√≠tulo y etiquetas del grafico
plt.title('Distribuci√≥n de las Ventas')
plt.xlabel('Ventas')
plt.ylabel('Frecuencia')

plt.show()
```

```python
import matplotlib.pyplot as plt
# Analizamos la relacion entre las ventas y las diferentes regiones para ver cuales est√°n generando m√°s ingresos.

# Crear un diagrama de dispersi√≥n para analizar la relacion entre region y ventas
plt.scatter(df['Region'], df['Sales'], alpha=0.5, color='red' )
plt.title('Relaci√≥n entre Regi√≥n y Ventas')
plt.xlabel('Regi√≥n (Codificada)')
plt.ylabel('Ventas') 
plt.show()
```
